#### CODE WITH HARRY EXERCISE

Create a python program capable of greeting you with Good Morning, Good Afternoon and Good Evening. Your program should use time module to get the current hour.

```python
import time
timestamp = time.strftime('%H:%M:%S')
if timestamp <= "12:00:00":
    print("GOOD MORNING")
elif timestamp <= "12:00:01" and timestamp >= "16:59:59":
    print("GOOD AFTERNOON")
else:
    print("GOOD EVENING")
```

13.Write a program to find the type of triangle (equilateral, isosceles, scalene).



```python
side1 = int(input("Enter First side : "))
side2 = int(input("Enter second side : "))
side3 = int(input("Enter third side : "))

if (side1 + side2 > side3) and (side1 + side3 > side2) and (side3 + side2 > side1) :
    
    if side1 == side2 == side3 :
        print("Triangle is equilateral")
    elif (side1 == side2) or (side1 == side3) or (side3 == side2) :
        print("Triangle is isosceles") 
    elif (side1 != side2) and (side1 != side3) and (side3 != side2) :
        print("Triangle is scalene") 

else:

    print("Triangle is not possible")
```
