1 .Write a program using match-case to display the day name based on a number (1–7).

```python
day = int(input("Enter number from 1-7 according to days : "))

match day :
    case 1 :
        print("Sunday")
    case 2 :
        print("Monday")
    case 3 :
        print("Tuesday")
    case 4 :
        print("wednesday")
    case 5 :
        print("Thursday")
    case 6 :
        print("Friday")
    case 7 :
        print("Saturday")
    case _ :
        print("Enter valid NUmber ")
```

2.Write a menu-driven calculator program using match-case
(addition, subtraction, multiplication, division).

```python
num1  = int(input("ENter first number : "))
num2  = int(input("ENter second number : "))

operation = int(input("Enter 1-4 according to do operations like add , sub , mul and div :"))

match operation:
    case 1 :
        print("Addition = ", num1 + num2 )
    case 2 :
        print("Substraction = ", num1 - num2 )
    case 3 :
        print("Multiplication = ", num1   *  num2 )
    case 4 :
        print("Division = ", num1 / num2 )
    case _ :
        print("Enter valid choice")
```

3.Write a program using match-case to check whether a character is a vowel or a consonant.

```python
char = input("Enter A character : ")
vowels = ['A','a','E','e','I','i','O','o','U','u']

match char:
    case _ if char in vowels: 
        print(char, " is vowel")
    case _:
        print(char , " is consonant ")
```

4.Write a program using match-case to display the grade of a student based on grade input
(A, B, C, D, F).

```python
grade = input("Enter your Grade : ")

match grade :
    case "A" | "a":
        print("Grade A")
    case "B" | "b":
        print("Grade B")
    case "C" | "c":
        print("Grade C")
    case "D" | "d":
        print("Grade D")
    case "F" | "F":
        print("Grade F")
    
    case _ :
        print("Enter valid Grade")
```
